# Data Cleaning with Python

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\a\Downloads\Nashville Housing Data for Data Cleaning (2).xlsx")
```

```python
df
```

# Drop Duplicates

```python
df.drop_duplicates()
```

# Drop NaN Values

```python
df.dropna(inplace = True)
```

# Description of Data

```python
df.describe()
```

# Expanding PropertyAddress

```python
df['PropertyAddress']
```

```python
df[['Street', 'City']] = df['PropertyAddress'].str.split(',', expand = True)
```

```python
df
```

# Put all the prices in dollars

```python
df.head(2)
```

```python
df['SalePrice'] = df['SalePrice'].astype('float')
```

```python
# col_to_add = [df.iloc[:, 5], df.iloc[:, 12], df.iloc[:, 13], df.iloc[:, 14]]

salesprice = np.array(df['SalePrice'].values)
landvalue = np.array(df['LandValue'].values)
buildingvalue = np.array(df['BuildingValue'].values)
totalvalue = np.array(df['TotalValue'].values)

formatted_sales = [f'${num:,.2f}' for num in salesprice]
formatted_land = [f'${num:,.2f}' for num in landvalue]
formatted_building = [f'${num:,.2f}' for num in buildingvalue]
formatted_total = [f'${num:,.2f}' for num in totalvalue]
```

```python
#do this for the rest
df['TotalValue'] = formatted_total
df['SalePrice'] = formatted_sales
df['BuildingValue'] = formatted_building
df['LandValue'] = formatted_sales
```

# Split OwnerAddress

```python
df['OwnerAddress'].head(5)
```

```python
df[['OwnerStreet', 'OwnerCity', 'State']] = df['OwnerAddress'].str.split(',', expand = True)
```

```python
df.head(2)
```

# Change Yearbuilt, Bedrooms, Fullbath and Halfbath to integers

```python
df
```

```python
df.info()
```

```python
year = np.array(df['YearBuilt'], np.int64)
rooms = np.array(df['Bedrooms'], np.int64)
full = np.array(df['FullBath'], np.int64)
half = np.array(df['HalfBath'], np.int64)

df['YearBuilt'] = year
df['Bedrooms'] = rooms
df['FullBath'] = full
df['HalfBath'] = half
```

```python
df.info()
```

```python
df.head(2)
```

# Drop propertyAddress and OwnerAddress columns

```python
df.drop(columns = ['PropertyAddress', 'OwnerAddress'], inplace = True)
```
